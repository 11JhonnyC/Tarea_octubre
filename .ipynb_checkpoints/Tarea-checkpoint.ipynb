````python
p_liebres = data[0:21,1] #-------------------columna liebres
p_linces = data[0:21,2] #--------------------columna linces
p_zanahorias = data[0:21,3] #----------------columna zanahorias
liebre_lince = data[0:21,1] + data[0:21,2] #-columna liebre + lince
axis_x = data[:, 0]#-------------------------columna años
````

Ahora haremos la gráficas para cada poblción:

Como agregado, ahora graficaremos las tendencias de cada población en planos individuales, de tal forma que podamos ver más claramente la tendencia de cada una, añadiendo a esto una línea guia que nos ayudará a visualizar aun mejor lo anterior.



## tomado de : [How to Add a Trend Line to a Line Chart/Graph](https://dzone.com/articles/python-how-to-add-trend-line-to-line-chartgraph#)



Para lo anterior, lo que haremos será crear ``subplots`` le ordenamos que se grafique ahí cada población en función del tiempo y le ordenamos que se grafique también una lñinea guia de tendencia.

````python
#-----Acá le ponemos los titulos a los plots y creamos los subplots-----#
a = plt.figure(figsize = (18,10))
a.suptitle("Líneas de Tendencia - Gráficas individuales" , fontsize=30)

ax = a.add_subplot(221)
ay = a.add_subplot(222)
az = a.add_subplot(223)
aw = a.add_subplot(224)
ax.set_title('Liebres', fontsize=20)
ay.set_title('Linces', fontsize=20)
az.set_title('Zanahorias', fontsize=20)
aw.set_title('Liebres + Linces', fontsize=20)
    

#-----De acá en adelante programamos para que se grafique para cada población con su línea de tendencia correspondiente-----#
    
zx = np.polyfit(axis_x, p_liebres, 1) #--Gráfica con línea de tendendia para liebres--#
tx = np.poly1d(zx)                   #---el parámetro 1 indica la curvatura de la línea. 10 es el máximo---#
ax.plot(axis_x, tx(axis_x),"r--") 
ax.plot(axis_x, p_liebres, 'r', alpha = 0.7)
ax.set_xticks(range(1900,1922,2))


zy = np.polyfit(axis_x, p_linces, 1)
ty = np.poly1d(zy)
ay.plot(axis_x,ty(axis_x),"r--")
ay.plot(axis_x, p_linces, 'g')
ay.set_xticks(range(1900,1922,2))


zz = np.polyfit(axis_x, p_zanahorias, 1)
tz = np.poly1d(zz)
az.plot(axis_x,tz(axis_x),"r--")
az.plot(axis_x, p_zanahorias, 'b',alpha = 0.8)
az.set_xticks(range(1900,1922,2))


zw = np.polyfit(axis_x, liebre_lince, 1)
tw = np.poly1d(zw)
aw.plot(axis_x,tw(axis_x),"r--")
aw.plot(axis_x, liebre_lince, 'm')
aw.set_xticks(range(1900,1922,2))

a.text(0.51, 0.04, 'Años', ha='center', fontsize=25)
a.text(0.05, 0.5, 'Poblaciones', va='center', rotation='vertical', fontsize=30)
````

## Manipulacion de matrices

- ``min(<<columna/fila_cliente>>)`` ---> Para el mínimo.

- ``sum(<<columna/fila_cliente>>)`` ---> Para la suma.

Ahora haremos lo anterior para una matriz 20x20 que tome valores de -50 a 50.

````python
#Hacemos la matriz
A, B = 20, 20
a1 = np.random.randint(-50,50,400)
m = a1.reshape(A,B) #-----La matriz con la que trabajaremos se llamará 'm'----#
````

Luego definimos la función la cual trabajará con dos variables, una que será la desición del cliente (el querer ver una fila o una columna) y la otra que será el número de esa fila o columna.

### Nota:
En caso tal de que el cliente sea una persona alejada de la programación, es preciso recordar que Python cuenta desde 0, en decir que si desea analizar la primera columna debe de saber que esta tiene como índice 0.

````python
def matriz(c, z):

#-----En caso de que desee una fila----#    
    if c == 1:
        fil = m[z,:]
        print('El valor mínimo en la fila seleccionada es:', min(fil))
        print('El valor máximo en la fila seleccionada es:', max(fil))
        print('La suma de los elementos de la fila seleccionada es:', sum(fil))
        print('La suma cuadrática de los elementos de la fila seleccionada es:', np.sqrt(sum(fil**2)))

#-----En caso de que desee una columna----#        
    elif c == 2:
        col = m[:,z]
        print('El valor mínimo en la columna selecionada es:', min(col))
        print('El valor máximo en la columna selecionada es:', max(col))
        print('La suma de los elementos de la columna seleccionada es:', sum(col))
        print('La suma cuadrática de los elementos de la columna seleccionada es:', np.sqrt(sum(col**2)))  
        
 #-----En caso de que digite un número por fuera del rango----#       
    else:
        print('Cualquier otro número, diferente de 1 o 2, no será leído por el código.')
````

### Ahora sí definimos la función:

````python
#-----En caso de que desee una fila----#
def matriz_cliente(b):
    
    if b == 1:               #-----Matriz con elementos sólo en su diagonal----#
        m_diago = np.diag(np.diag(m))
        print(m_diago)
        
    elif b == 2:             #-----Matriz con elementos sólo en sus bordes----#
        m_bordes = m[1:20-1,1:20-1]=0  
        print(m)
        
    elif b == 3:             #-----Suma de los valores absolutos de los elementos de la matriz----#
        sum_abs = sum(sum(np.absolute(m[:,:])))
        print(sum_abs)
        
    elif b == 4:              #-----Suma de los máximos de cada fila de la matriz----#
        maximos = [] #--Lista donde se guardarán los máximos--#
        indice=0
        for i in m:
            q = max(m[indice,:])
            maxs = maximos.append(q)
            indice=indice+1  
        print(sum(maximos))
        
    else :
        print('Cualquier otro número no será leída por el código. Asegurese de que el número digitado se encuentre en las opciones.')
````
